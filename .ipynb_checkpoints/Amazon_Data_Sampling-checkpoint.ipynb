```python
start_brand = pd.DataFrame(columns=amazon_df.columns)
middle_brand = pd.DataFrame(columns=amazon_df.columns)
end_brand = pd.DataFrame(columns=amazon_df.columns)
# Getting the count of rows having the brand in the beginning, end and middle
for i,row in amazon_df.iterrows():
  brand = row.brand
  #print(brand)
  title = row.title
  if title[:len(brand)] == brand:
    start_brand = start_brand.append(row)
  elif title[-len(brand):-1] == brand:
    end_brand = end_brand.append(row)
  else:
    middle_brand = middle_brand.append(row)
```

```python
unique_start_brand = start_brand["brand"].value_counts() 
unique_start_brand
```

```python
groupped_start_brand = start_brand.groupby("brand")
```

```python
new_start_brand = groupped_start_brand.filter(lambda x: len(x) < 3)
```

```python
tmp_grp_start_brand = groupped_start_brand.filter(lambda x: len(x) >= 3)
tmp_grp_start_brand = tmp_grp_start_brand.groupby("brand")
```

```python
to_sample = pd.DataFrame(columns = new_start_brand.columns)
for g_idx, group in tmp_grp_start_brand:
  counter = 0
  half_len = int(group.shape[0]/2)
  
  for r_idx, row in group.iterrows(): 
    counter+=1 
    if counter <= half_len :
      new_start_brand = new_start_brand.append(row)
    else:
      to_sample = to_sample.append(row)
```

```python
new_start_brand = new_start_brand.reset_index()
```
